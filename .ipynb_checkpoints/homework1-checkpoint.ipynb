* Standardize columns `age`, `est_income` and `local`.
* Derive five features by applying a PCA. Which of the original features has the highest correlation with one of the five derived features (principal components)?

```python
# excersie1
import pandas as pd

# Aufgabe 1
df = pd.read_excel('./data/customers.xls')

# Aufgabe 2
d = {'M':0,'F':1,'S':1, 'D':2,'N':0,'Y':1,'CC':0,'CH':1,'Auto':2,'Budget':0,'FreeLocal':1,'Intnl_discount':0,'Standard':1,'Cancelled':0,'Current':1}
df = df.replace(d)

print(df)

# Aufgabe 3
from sklearn.preprocessing import StandardScaler

df = pd.read_excel('./data/customers.xls')
df_changing = df[['Age', 'Est_Income', 'Local']]

standardscaler = StandardScaler();
df_standardized = standardscaler.fit_transform(df_changing)

df_standardized

df_new = pd.DataFrame({'Age':df_standardized[0:,0],'Est_Income':df_standardized[0:,1],'Local':df_standardized[0:,2]})

print(df_new)

# Aufgabe 4
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

df = pd.read_excel('./data/customers.xls')
d = {'M':0,'F':1,'S':1, 'D':2,'N':0,'Y':1,'CC':0,'CH':1,'Auto':2,'Budget':0,'FreeLocal':1,'Intnl_discount':0,'Standard':1,'Cancelled':0,'Current':1}
df = df.replace(d)

standardscaler2 = StandardScaler();
df_standardized2 = standardscaler2.fit_transform(df)

pca = PCA(n_components=5)

pca.fit(df_standardized2)


# looking for correlation of original components and principal components

print(pca.components_)


```
